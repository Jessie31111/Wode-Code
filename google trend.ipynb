```python
def get_value(country, keyword, start_time_str, end_time_str):
    # url to get data info and token
    url = 'https://trends.google.com/trends/api/explore?hl=zh-HK&tz=-480&req={"comparisonItem":[{"keyword":"'+keyword+'","geo":"'+country+'","time":"'+start_time_str+'+'+end_time_str+'"}],"category":0,"property":""}&tz=-480'
    r = requests.get(url, headers=headers, timeout=15)
    data = json.loads(r.text[5:])
    req = data['widgets'][0]['request']
    token = data['widgets'][0]['token']
    
    # url to get daily data
    url2 = f'https://trends.google.com/trends/api/widgetdata/multiline?hl=zh-CN&tz=-480&req={req}&token={token}&tz=-480'
    r2 = requests.get(url2)

    if r2.status_code == 200: # if status code is 200, then get data
        data = json.loads(r2.text.encode().decode('unicode_escape')[6:])['default']['timelineData']
        date_list = []
        value_list = []
        df_value = pd.DataFrame(columns=['date', 'value', 'country', 'keyword'])
        for data_e in data:
            timestampstr = time.strftime("%Y-%m-%d", time.localtime(int(data_e['time'])))
            valuenum = data_e['value'][0]
            date_list.append(timestampstr)
            value_list.append(valuenum)
        df_value['date'] = date_list
        df_value['value'] = value_list
        df_value['country'] = country
        df_value['keyword'] = keyword   
        
    else:
        print("error")
    
    return df_value
```
